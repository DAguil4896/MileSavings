```python
from mysql.connector import MySQLConnection, Error

def iter_row(cursor, size=50000):
    while True:
        rows = cursor.fetchmany(size)
        if not rows:
            break
        for row in rows:
            yield row

def test_fetchmany():
    conn = MySQLConnection(host='Dannys-MacBook-Pro.local', database='FinalProject', user='root', password='Aguilard123')
    cursor = conn.cursor()

    cursor.execute("SELECT * FROM FinalSequence1FirstLessThanSecond ")
    queryList = []
    counter = 0
    for row in iter_row(cursor, 50000):
        queryList.append(row)
        if counter >= 50000:
               break
        counter = counter + 1
        
    
    return queryList
        
```

```python
listJan1 = []
listJan1 = test_fetchmany()
jan1 = pd.DataFrame(listJan1)
jan1.columns = ['Sequence1Dist', 'Sequence2Dist', 'Sequence3Dist', 'Sequence4Dist', 'TotalDistanceP2P', 'checkSequence1Timing']
jan1
```

```python
#Creates Empty lists to store sequences
myList11 = []
myList12 = []
myList13 = []
myList14 = []

#Loops through to check whoch sequence gives the best saving
i = 0
for row in jan1.itertuples():
    if(jan1.at[i, 'TotalDistanceP2P'] > jan1.at[i, 'Sequence1Dist']):
        myList11.append(row)
    elif(jan1.at[i, 'TotalDistanceP2P'] > jan1.at[i, 'Sequence2Dist']):
        myList12.append(row)
    elif (jan1.at[i, 'TotalDistanceP2P'] > jan1.at[i, 'Sequence3Dist']):
        myList13.append(row)
    elif (jan1.at[i, 'TotalDistanceP2P'] > jan1.at[i, 'Sequence4Dist']):
        myList14.append(row)

    i = i + 1


#Convert the lists into DataFrame 
jan1DF1 = pd.DataFrame(myList11)
jan1DF2 = pd.DataFrame(myList12)
jan1DF3 = pd.DataFrame(myList13)
jan1DF4 = pd.DataFrame(myList14)

#Calculate the savings in a new column
for row in jan1DF1.itertuples():
    jan1DF1["Savings1"] = jan1DF1["TotalDistanceP2P"] - jan1DF1["Sequence1Dist"]


#Display the dataframes    
jan1DF1
```

```python
#Creates Empty lists to store sequences
myList11 = []
myList12 = []
myList13 = []
myList14 = []

#Loops through to check whoch sequence gives the best saving
i = 0
for row in yMay4.itertuples():
    if(yMay4.at[i, 'TotalDistanceP2P'] > yMay4.at[i, 'Sequence1Dist']):
        myList11.append(row)
    elif(yMay4.at[i, 'TotalDistanceP2P'] > yMay4.at[i, 'Sequence2Dist']):
        myList12.append(row)
    elif (yMay4.at[i, 'TotalDistanceP2P'] > yMay4.at[i, 'Sequence3Dist']):
        myList13.append(row)
    elif (yMay4.at[i, 'TotalDistanceP2P'] > yMay4.at[i, 'Sequence4Dist']):
        myList14.append(row)

    i = i + 1


#Convert the lists into DataFrame 
yMay4DF4 = pd.DataFrame(myList14)

#Calculate the savings in a new column
for row in yMay4DF4.itertuples():
    yMay4DF4["Savings4"] = yMay4DF4["TotalDistanceP2P"] - yMay4DF4["Sequence4Dist"]

#Display the dataframe    
yMay4DF4
```

```python
#Calculate the total savings
yMay4DF4['Savings4'].sum()
```
